```python
import numpy as np
import gym
import random
```

```python
env = gym.make('FrozenLake-v1',is_slippery=False)
```

```python
qtable=np.zeros((16,4))
qtable
```

```python
state=env.reset()
qtable[state]
```

```python
def rargmax(vector):    
    m = np.amax(vector)

    indices = np.nonzero(vector == m)

    return random.choice(indices[0])

action_test = rargmax([0,0,1,0])
action_test
```

```python
# RL training begins


# saves the new reward value of an episode (+1)
rList = []

for i in range(1000):
    state = env.reset()
    rAll = 0
    done = False
    sequence = []

    while not done:
        env.render()

        action = rargmax(qtable[state,:])
        #action = random.randint(1,2)
        
        sequence.append(action)
        new_state, reward, done, info = env.step(action)
        
        # Update table
        qtable[state, action] = reward + np.max(qtable[new_state, :])
        state = new_state

        rAll += reward

    rList.append(rAll)
    print(f"Sequence of an episode = {sequence}")

env.close()
```

```python
rList
```

```python
Score = sum(rList)/1000
Score
```
