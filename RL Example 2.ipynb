```python
import numpy as np
import gym
import random
```

```python
env = gym.make('FrozenLake-v1',is_slippery=False)
```

```python
qtable = np.zeros((env.observation_space.n, env.action_space.n))
```

```python
# Hyperparameters
alpha = 0.5           
gamma = .9         

def rargmax(vector):   
    m = np.amax(vector)

    indices = np.nonzero(vector == m)

    return random.choice(indices[0])

rList = []

for i in range(1000):
    state = env.reset()
    rAll = 0
    done = False

    while not done:
        env.render()
    
        action = rargmax(qtable[state,:])
        new_state, reward, done, info = env.step(action)
        
        # Update table
        qtable[state,action] = (1-alpha) * qtable[state,action] \
           + alpha*(reward + gamma * np.max(qtable[new_state,:]))
                
        rAll += reward
        state = new_state

    rList.append(rAll)
```
