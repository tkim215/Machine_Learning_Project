```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
original_df = pd.read_csv("C:/Users/kimta/Introduction to Machine Learning/raw_data/Social_Network_Ads.csv")
original_df
```

```python
df = original_df[['Age', 'EstimatedSalary']]
df
```

```python
plt.boxplot(df)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
df_scaled=scaler.transform(df)
pd.DataFrame(df_scaled, columns=["age_std","estSalary_std"])
```

```python
print(df.mean())
```

```python
plt.boxplot(df_scaled)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_minmax = scaler.fit_transform(df)
pd.DataFrame(df_minmax, columns = ['age_minmax', 'estSalary_minmax'])
```

```python
plt.boxplot(df_minmax)
plt.show()
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
df_robust = scaler.fit_transform(df)
pd.DataFrame(df_robust, columns = ['age_robust', 'estSalary_robust'])
```

```python
plt.boxplot(df_robust)
plt.show()
```
