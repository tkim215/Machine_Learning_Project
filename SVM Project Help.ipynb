```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/kimta/Introduction to Machine Learning/raw_data/bcancer.csv")
```

```python
df.columns
```

```python
sns.pairplot(df, hue='diagnosis', vars=['texture_mean','area_mean','radius_mean'])
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
target = LE.fit_transform(df['diagnosis'])

df.drop('diagnosis', axis=1, inplace=True)
df['diagnosis'] = target
df
```

```python
X=df.drop(['diagnosis','id','radius_mean'],axis=1)
X
```

```python
y=df['diagnosis']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear', C=1)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])

sns.heatmap(confusion_matrix, annot=True)
plt.show()

# feature = dimension 
```
